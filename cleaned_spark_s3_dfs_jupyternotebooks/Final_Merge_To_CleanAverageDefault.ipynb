```python
import pandas as pd
```

```python
final_merge_df = pd.read_csv("../Resources/Final_Merge2.csv")
final_merge_df.head()
```

```python
#list the columns
final_merge_df.columns
```

```python
# Get the distinct NPA values
npa_values = final_merge_df['NPA'].unique()
npa_values
```

```python
# for each npa value, get the column values that are numeric
numeric_columns = ['Acres', 'births_to_adolescents', 'grocery_proximate_units', 'prenatal_care', 'grocery_proximity',
                  'low_birthweight', 'public_health_insurance', 'age_of_death', 'impervious_surface',
                  'water_consumption', 'neighborhood_school_attendance', 'highschool_graduation_rate', 
                   'proficiency_elementary_school', 'proficiency_middle_school', 'proficiency_high_school',
                   'student_absenteeism', 'voters_participating', 'neighborhood_organizations',
                   'voter_participation', 'vacant_land', 'population_density', 'population', 'transit_ridership',
                   'street_connectivity', 'bicycle_friendliness'
]
for npa_value in npa_values:
    for numeric_column in numeric_columns:
        npa_df = final_merge_df.loc[final_merge_df['NPA'] == npa_value]
        npa_numeric_df = npa_df[numeric_column]
        #print(npa_numeric_df.head())
        mean_value = npa_numeric_df.mean()
        print(mean_value)
        avg_df = npa_numeric_df.fillna(mean_value)
        avg_df = avg_df.fillna(-1)
        npa_df[numeric_column] = avg_df
        
        # replace NaN values with -999
        #final_merge_df.loc[(final_merge_df['NPA'] == npa_value),numeric_column].fillna(-9999, inplace=True)
        #final_merge_df.loc[((final_merge_df['NPA'] == npa_value) & (final_merge_df[numeric_column] == -999)),numeric_column]= mean_value

        final_merge_df.loc[(final_merge_df['NPA'] == npa_value),numeric_column] = avg_df
        print(final_merge_df.loc[(final_merge_df['NPA'] == npa_value),numeric_column].head())
        print(avg_df)
        #print(npa_df.head())
```
